# Xception Keras Pretrained Model

```python
import tensorflow as tf
from tensorflow.keras.applications import Xception
from tensorflow.keras.applications.xception import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
```

### Model Loading

```python
# Load the pretrained Xception model
model = Xception(weights='imagenet')
```

```python
model.summary()
```

```python
'''GAP reduces the number of parameters by over 98% in the transition to the final layer. This makes your model much smaller and faster to train.
latten creates millions of weights, the model becomes "too powerful." It starts memorizing the exact position of pixels in your training images'''
```

```python
# Prediction Function
def predict_image(model, image_path):
    img = load_img(image_path, target_size=(299, 299))  # Must be 299x299
    img_array = img_to_array(img) # image obj to numpy array
    img_array = np.expand_dims(img_array, axis=0) # batch dim added tesnflow work with extra dim
    img_array = preprocess_input(img_array)

    preds = model.predict(img_array)
    return decode_predictions(preds, top=3)[0]

```

```python
import requests

def download_image(url, save_path):
  """
  Downloads an image from a given URL and saves it to a specified path.

  Args:
      url: The URL of the image.
      save_path: The path where the image will be saved.
  """
  try:
    response = requests.get(url, stream=True)
    response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
    with open(save_path, 'wb') as out_file:
      for chunk in response.iter_content(chunk_size=8192):
        out_file.write(chunk)
    print(f"Image downloaded successfully to {save_path}")
  except requests.exceptions.RequestException as e:
    print(f"Error downloading image: {e}")
```

```python
download_image("https://cdn.mos.cms.futurecdn.net/HjFE8NKWuCmgfHCcndJ3rK.jpg", "zebra.jpg")
```

```python
image_path = r"C:\Users\Jagdish singh\DEEP-LEARNING\Transfer_Learning_img_classfxn\zebra.jpg"
predictions = predict_image(model, image_path)

# Display the predictions
print("Top Predictions:")
for class_id, description, score in predictions:
    print(f"{description}: {score:.4f}")

# Display the image
plt.imshow(load_img(image_path, target_size=(299, 299)))
plt.axis("off")
plt.title("Input Image")
plt.show()
```
