```python
import requests

def download_image_from_url(url, save_path):
  """
  Downloads an image from a given URL and saves it to a local path.

  Args:
    url (str): The URL of the image.
    save_path (str): The local path to save the image to.
  """
  try:
    response = requests.get(url, stream=True)
    response.raise_for_status()  # Raise an exception for bad status codes

    with open(save_path, 'wb') as out_file:
      out_file.write(response.content)
    print(f"Image downloaded successfully to {save_path}")
  except requests.exceptions.RequestException as e:
    print(f"Error downloading image: {e}")
```

```python
import requests
from IPython.display import Image, display

def download_image(url, save_path):
    # This header makes the request look like it's coming from a real browser
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
    }
    try:
        response = requests.get(url, stream=True, headers=headers)
        response.raise_for_status()
        with open(save_path, 'wb') as out_file:
            for chunk in response.iter_content(chunk_size=8192):
                out_file.write(chunk)
        print(f"✅ Image downloaded successfully to {save_path}")
    except Exception as e:
        print(f"❌ Error downloading image: {e}")

# Try the download again
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/de/Davidraju_Vine_snake.jpg/960px-Davidraju_Vine_snake.jpg"
save_name = "vine_snake.jpg"

download_image(url, save_name)

# Display if download was successful
try:
    display(Image(filename=save_name))
except:
    print("Could not display image because download failed.")
```
